Task 2:
Write a function to calculate average, median, and
mode of a list of numbers

```python
import statistics

def calculate_stats(numbers):
  """Calculates the average, median, and mode of a list of numbers.

  Args:
    numbers: A list of numbers.

  Returns:
    A dictionary containing the average, median, and mode of the list.
  """
  average = statistics.mean(numbers)
  median = statistics.median(numbers)
  try:
    mode = statistics.mode(numbers)
  except statistics.StatisticsError:
    mode = "No unique mode"  # Handle cases with no unique mode

  return {"average": average, "median": median, "mode": mode}

# Example usage:
data = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]
stats = calculate_stats(data)
print(f"The statistics for the list {data} are: {stats}")

# Example with no unique mode:
data_no_mode = [1, 2, 3, 4, 5, 6]
stats_no_mode = calculate_stats(data_no_mode)
print(f"The statistics for the list {data_no_mode} are: {stats_no_mode}")
```

```python
def convert_to_binary(num):
  """Converts a decimal number to its binary representation.

  Args:
    num: An integer representing the decimal number.

  Returns:
    A string representing the binary representation of the number.
  """
  return bin(num).replace("0b", "")

# Example 1
input_1 = 10
output_1 = convert_to_binary(input_1)
print(f"Input: {input_1}, Output: {output_1}")

# Example 2
input_2 = 25
output_2 = convert_to_binary(input_2)
print(f"Input: {input_2}, Output: {output_2}")

# Example 3
input_3 = 7
output_3 = convert_to_binary(input_3)
print(f"Input: {input_3}, Output: {output_3}")

# Example 4
input_4 = 0
output_4 = convert_to_binary(input_4)
print(f"Input: {input_4}, Output: {output_4}")
```

```python
def celsius_to_fahrenheit(celsius):
  """Converts a temperature from Celsius to Fahrenheit.

  Args:
    celsius: A float or integer representing the temperature in Celsius.

  Returns:
    A float representing the temperature in Fahrenheit.

  Raises:
    TypeError: If the input `celsius` is not a number (float or int).

  Example:
    Input: 0 (Celsius)
    Expected Output: 32.0 (Fahrenheit)
  """
  if not isinstance(celsius, (int, float)):
    raise TypeError("Input must be a number (integer or float).")
  return (celsius * 9/5) + 32

# Example usage:
print(f"0 Celsius is: {celsius_to_fahrenheit(0)} Fahrenheit")
print(f"25 Celsius is: {celsius_to_fahrenheit(25)} Fahrenheit")
```
